```python
import pandas as pd # import library 
import numpy as np
```

```python
dataset= pd.read_csv("insurance.csv")  # function of pandas used to load the data
dataset
```

```python
dataset=pd.get_dummies(dataset) #to create dummies value for categorical text
dataset
```

```python
dataset=dataset.replace({True:1,False:0}) #replce the 0 and 1 for true,false
dataset
```

```python
dataset = dataset.drop(columns=["sex_female", "smoker_no"])# drop columns from the dataset
dataset
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]  # split input and output to get input from dataset 
independent
```

```python
dependent=dataset[['charges']] # get output from dataset
dependent
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV # built model for grid search and cross validation
param_grid={'random_state':[10,12,15],'n_estimators':[100,200,10],'learning_rate':[2.0,1.0],'loss':['linear']}
grid= GridSearchCV( AdaBoostRegressor(),param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_ # print best parameter after tuning
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
import pickle   # store the model
filename = "finalized_model_ada_BOOST_grid.sav" 
pickle.dump(grid,open(filename,"wb"))
```

```python
load_model=pickle.load(open("finalized_model_ada_BOOST_grid.sav" ,"rb")) # load the model from the file
result=load_model.predict([[50,30.970,3,1,0]])
```
