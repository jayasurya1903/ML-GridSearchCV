```python
import pandas as pd   # it's a library used process the tubular data
```

```python
dataset=pd.read_csv("50_Startups.csv")   # function of pandas used to load the data
dataset
```

```python
dataset=pd.get_dummies(dataset)  #to create dummies value for categorical text
dataset
```

```python
dataset = dataset.replace({True: 1, False: 0}) #replce the 0 and 1 for true,false
dataset
```

```python
dataset = dataset.drop(columns=['State_California']) # drop first column from the dataset
dataset
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']] # split input and output to get input from dataset
independent
```

```python
dependent=dataset[['Profit']] # get output from dataset
dependent
```

```python
from sklearn.model_selection import GridSearchCV # built model for grid search and cross validation
from sklearn.ensemble import RandomForestRegressor
param_grid={'criterion':['squared_error','friedman_mse','absolute_error','poisson'],
            'n_estimators':[10,50,200,30],'max_depth':[None],'random_state':[None]}
grid= GridSearchCV(RandomForestRegressor(),param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_ # print best parameter after tuning
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
import pickle   # store the model
filename = "finalized_model_random_forest_grid_regression.sav" 
pickle.dump(grid,open(filename,"wb"))
```

```python
load_model=pickle.load(open("finalized_model_random_forest_grid_regression.sav" ,"rb")) # load the model from the file
result=load_model.predict([[153441.51,101145.55,407934.54,1,0]])
```
